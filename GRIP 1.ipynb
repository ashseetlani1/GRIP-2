## Importing All the libraries we will need in this task.

##### Here we imported Numpy for Arrays, Matplotlib for plotting Graphs, Pandas for DataFrames,KMeans from sklearn.cluster to perform KMeans Clustering.

```python
from pandas import read_csv
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

##### Here we imported the dataset and displayed first 5 rows of it.

```python
dataframe = read_csv('C:\Iris.csv')
print(dataframe.head(5))
```

##### Here we dropped the result table and framed the inputs into x

```python
x = dataframe.iloc[:,[1,2,3,4]].values
print(x[:5,:])
```

##### Then we fit the data into KMeasn algorithm.

```python
kmeans = KMeans(3)
y_kmeans = kmeans.fit_predict(x)
```

##### As you see we fit the data into algorithm, and now we plotting clusters.

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
plt.show()
%matplotlib inline
```
